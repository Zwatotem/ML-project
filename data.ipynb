```python
import pandas as pd
import numpy as np
```

```python
events = pd.read_csv('events.csv')
events.head()
```

```python
events.columns
```

```python
players1 = events['player']
players2 = events['player2']
playersin = events['player_in']
playersout = events['player_out']
players = pd.concat([players1, players2, playersin, playersout], ignore_index=True)
players_tally = players.value_counts()
players_tally = pd.DataFrame(players_tally)
players_tally['player'] = players_tally.index
players_tally['index'] = range(1, len(players_tally)+1)
players_tally.head()
players_replacement_dict = players_tally.set_index('player')['index'].to_dict()
players_replacement_dict
```

```python
events.replace({'player': players_replacement_dict}, inplace=True)
events.replace({'player2': players_replacement_dict}, inplace=True)
events.replace({'player_in': players_replacement_dict}, inplace=True)
events.replace({'player_out': players_replacement_dict}, inplace=True)
# events['id_event','player','player2','player_in','player_out'].head()
```

```python
events[['id_event','event_type','player','player2','player_in','player_out']][events.event_type==7].head()
```

```python
teams1 = events['event_team']
teams2 = events['opponent']
teams = pd.concat([teams1, teams2], ignore_index=True)
teams_tally = teams.value_counts()
teams_tally = pd.DataFrame(teams_tally)
teams_tally['team'] = teams_tally.index
teams_tally['index'] = range(1, len(teams_tally)+1)
teams_tally.head()
teams_replacement_dict = teams_tally.set_index('team')['index'].to_dict()
teams_replacement_dict
```

```python
events.replace({'event_team': teams_replacement_dict}, inplace=True)
events.replace({'opponent': teams_replacement_dict}, inplace=True)
events.head()[['id_event','event_type','event_team','opponent']]
```

```python
events['open_play'] = 1*(events.situation == 1)
events['set_piece'] = 1*(events.situation == 2)
events['corner'] = 1*(events.situation == 3)
events['free_kick'] = 1*(events.situation == 4)
events.head()
```

```python
events.drop(['situation'], axis=1, inplace=True)
events.head()
```

```python
events.isnull().sum()
```

```python
commonst_location = events.location.mode()[0]
events.location.fillna(commonst_location, inplace=True)
```

```python
events.drop(['shot_outcome'], axis=1, inplace=True)
```

```python
matches = events.groupby(['id_odsp'])
timesteps = 0
for name, group in matches:
    timesteps = max(timesteps, len(group))
timesteps
```

```python
group_list = []
for name, group in matches:
    group['is_team1_goal'] = group['is_goal'].__and__(group['event_team'] == min(group.event_team))
    group['is_team2_goal'] = group['is_goal'].__and__(group['event_team'] == max(group.event_team))
    group['team1_cumgoals'] = group['is_team1_goal'].cumsum()
    group['team2_cumgoals'] = group['is_team2_goal'].cumsum()
    group.drop(['is_team1_goal', 'is_team2_goal'], axis=1)
    group_list.append(group)
```

```python
X_list = []
y_list = []
for name, group in matches:
    group['is_team1_goal'] = group['is_goal'].__and__(group['event_team'] == min(group.event_team))
    group['is_team2_goal'] = group['is_goal'].__and__(group['event_team'] == max(group.event_team))
    group['team1_cumgoals'] = group['is_team1_goal'].cumsum()
    group['team2_cumgoals'] = group['is_team2_goal'].cumsum()
    group.drop(['is_team1_goal', 'is_team2_goal'], axis=1)
    X_list.append(group.drop(['id_odsp', 'team1_cumgoals','team2_cumgoals', 'is_goal' ], axis=1).to_numpy().tolist())
    y_list.append(group[['team1_cumgoals','team2_cumgoals']].to_numpy().tolist())
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
X_data = np.array(pad_sequences(X_list, maxlen=timesteps, padding='post'))
y_data = np.array(pad_sequences(y_list, maxlen=timesteps, padding='post'))
```
